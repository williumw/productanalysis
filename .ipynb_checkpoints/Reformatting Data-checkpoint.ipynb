```python
import pandas as pd
import numpy as np
import re
from pandas import ExcelWriter
```

```python
main = pd.read_excel('/Users/William/Desktop/codebase/timeseries clustering/sales_data_sample.xlsx')
main.head()
```

```python
#function objective: to figure out the products in order of sales amount from sample dataset

def top_product(data):
    df = data
    df = df.sort_values(by = 'Sales', ascending = False)
    df = df.reset_index(drop = True)
    
    master_product_number_list = []
    master_product_line = []
    sums = []
    
    for index, row in df.iterrows():
        if row['Product Code'] not in master_product_number_list and row['Product Line'] not in master_product_line:
            master_product_number_list.append(row['Product Code'])
            master_product_line.append(row['Product Line'])
    
    for product in master_product_number_list:
        sums.append(df.loc[df['Product Code'] == product, 'Sales'].sum())
        
    master_product_number_list.insert(0, 'Product Code')
    master_product_line.insert(0, 'Product Line')
    sums.insert(0, 'Sum of Sales')
    
    xd = pd.DataFrame({'Product Line': master_product_line, 'Product Number': master_product_number_list, 'Sum of Sales': sums})
    xd = xd.ix[1:]
    xd = xd.sort_values(by = 'Sum of Sales', ascending = False)
    xd = xd.reset_index(drop = True)
    return xd

top_product(main)
```

```python
def customer_names(data):
    df = data
    customer_list = []
    for index, row in df.iterrows():
        if row['Customer Name'] not in customer_list:
            customer_list.append(row['Customer Name'])
            
    return customer_list
customer_names(main)
```

```python
def product_year_quarter_list(data):
    df = data
    products = top_product(data)['Product Line'].tolist()
    name_year_quarter = []
    
    for product in products:
        for index, row in df.iterrows():
            if row['Product Line'] == product:
                name = '_'.join([row['Product Line'], str(row['Year']), str(row['Quarter'])])
                if name not in name_year_quarter:
                    name_year_quarter.append(name)
    return name_year_quarter

product_year_quarter_list(main)
```

```python
#function objective: to create a DataFrame that has the products as columns with their respective year and quarter and the accounts as the Y-axis to hold the quantity data

def complete_data(data):
    top_df = top_product(data)
    customer_list = customer_names(data)
    df = data
    name_year = []
    header = product_year_quarter_list(data)
    
    timeseries = pd.DataFrame(index = customer_list, columns = header)
    timeseries = timeseries.fillna(0)
    
    
    
    return timeseries

complete_data(main)
```
